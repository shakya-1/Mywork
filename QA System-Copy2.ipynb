```python
###############################################################################
############ Importing necessary library and downloading the model ############

import warnings
warnings.simplefilter('ignore')
import pandas as pd
from cdqa.utils.download import download_model
from cdqa.pipeline.cdqa_sklearn import QAPipeline
```

```python
###############################################################################
################# Reading of the file from directory ##########################

test=pd.read_csv('/home/rahul/Downloads/covid.csv',encoding='latin')
test
```

```python
###############################################################################
################### Converting the Answers in the list ########################

ans=[]
for i in range(len(test["Answers"])):
    ans.append([])
    ans[i].append(test["Answers"][i])
    test["Answers"][i]=ans[i]
test
```

```python
################################################################################
################# Renaming 'Answers' column to 'paragraphs' ####################
############# And adding a new column 'title' to the dataframe #################

test=test.rename(columns={"Answers":"paragraphs"})
test["title"]="Covid Question and Answer"
test
```

```python
##################################################################################
### Loading QAPipeline with CPU version of BERT Reader pretrained on SQuAD 1.1 ###

cdqa_pipeline = QAPipeline(reader='/home/rahul/models/bert_qa.joblib')
```

```python
##################################################################################
######### Fitting the retriever to the list of documents in the dataframe ########

cdqa_pipeline.fit_retriever(test)
```

```python
####################################################################################
############ Sending a question to the pipeline and getting prediction #############

query = input("Question: ")
prediction = cdqa_pipeline.predict(query)

if(prediction[0]!=''):
    print('Answer: {}\n'.format(prediction[0]))
    #print('title: {}\n'.format(prediction[1]))
    #print('paragraph: {}\n'.format(prediction[2]))
else:
    print("I don't have answer for that question.")
```
